```python
numpyArray = np.linspace(0,10,10)
```

```python
numpyArray2 = numpyArray **2
```

```python
plt.plot(numpyArray,numpyArray2,'g')
plt.show()
```

```python
plt.plot(numpyArray,numpyArray2,'g--')
plt.show()
```

```python
plt.plot(numpyArray,numpyArray2,'g*-')
plt.show()
```

```python
plt.plot(numpyArray,numpyArray2,'g*')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray,numpyArray2,'r*-')
plt.subplot(1,2,2)
plt.plot(numpyArray2,numpyArray,'b*-')
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.1,0.1,0.9,0.9])
figureAxes.plot(numpyArray,numpyArray2,'r')
figureAxes.set_xlabel('X Axes')
figureAxes.set_ylabel('Y Axes')
figureAxes.set_title('Title')
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyArray,numpyArray2,'g')
eksen1.set_xlabel('X Ekseni')
eksen1.set_ylabel('Y Ekseni')
eksen1.set_title('Main Graphic Title')

eksen2.plot(numpyArray2,numpyArray,'b')
eksen2.set_xlabel('X Ekseni')
eksen2.set_ylabel('Y Ekseni')
eksen2.set_title('Second Graphic Title')
```

```python
myFigure,myAxis = plt.subplots(nrows=1,ncols=2)
for axis in myAxis:
    axis.plot(numpyArray,numpyArray2,'b')
    axis.set_xlabel('X Label')
plt.tight_layout()
```

```python
newFigure = plt.figure(figsize=(5,5),dpi=70)

newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxis.plot(numpyArray,numpyArray ** 2,'r',label= "nunpyArray ** 2 ")
newAxis.plot(numpyArray,numpyArray ** 3,'b',label="numpyArray ** 3")
newAxis.set_xlabel('X Label')
newAxis.set_ylabel('Y Label')
newAxis.legend(loc=6)
```
