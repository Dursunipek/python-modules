# --- Matplotlib ---

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Ä°ntroduction

```python
age = np.arange(10,101,10)
weight = np.random.randint(50,120,10)
```

```python
print(weight)
print(age)
```

```python
plt.plot(age,weight,'r')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.show()
```

```python
numpyArray = np.linspace(0,100,20)
```

```python
numpyArray2 = numpyArray **2
```

```python
plt.plot(numpyArray,numpyArray2,'g')
plt.show()
```

```python
plt.plot(numpyArray,numpyArray2,'g--')
plt.show()
```

```python
plt.plot(numpyArray,numpyArray2,'g*-')
plt.show()
```

```python
plt.plot(numpyArray,numpyArray2,'g*')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray,numpyArray2,'r*-')
plt.subplot(1,2,2)
plt.plot(numpyArray2,numpyArray,'b*-')
```

```python
myFigure = plt.figure()
figureAxex = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxex.plot(numpyArray,numpyArray2,'r')
figureAxex.set_xlabel('X Axes')
figureAxex.set_ylabel('Y Axes')
figureAxex.set_title('Title')
```
